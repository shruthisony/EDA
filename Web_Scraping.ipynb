```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
try:
  l=[]

  my_url = requests.get('https://books.toscrape.com/')
  my_url.raise_for_status()

  soup = BeautifulSoup(my_url.text,'html.parser')

  Books = soup.find('ol',class_='row').find_all('li')
  #print(len(Books))   --20 in one page

  for i in Books:
    name = i.find('article',class_='product_pod').find('h3').find('a',href=True).get('title')
    price = i.find('article',class_='product_pod').find('div',class_='product_price').p.text
    available = i.find('article',class_='product_pod').find('div',class_='product_price').find('p',class_='instock availability').get_text(strip=True)
    rating = i.find('article',class_='product_pod').find('p').get('class')[1]
    print(name+','+price+','+available+','+rating)
    l.append([name,price,available,rating])

except Exception as e:
  print(e)
```

```python
print(l)
```

```python
df = pd.DataFrame(l)
df
```

```python
df.columns=['Name','Price','Available','Rating']
```
